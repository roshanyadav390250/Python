```python
def factorialnumber(n):
    if n==1 :
        return 1
    else:
        return n *factorialnumber(n-1)
```

```python
factorialnumber(5)
```

```python
def fabnocci(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return fabnocci(n-1)+ fabnocci(n-2)
```

```python
fabnocci(5)
```

```python
def fact(n):
    if n==1:
        return 1
    return n * fact(n-1)
```

```python
fact(5)
```

```python
def fab(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return fab(n-1)+fab(n-2)
```

```python
fab(10)
```
