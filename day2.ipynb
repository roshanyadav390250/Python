- In Python, a list is a built-in data type used to store multiple items in a single variable. Lists are ordered, mutable (changeable), and allow duplicate values. They are one of the most commonly used data structures in Python

```python
number=[1,2,3,4,5,6,7,8,9]
print(number)
```

## tuple ##
- In Python, a tuple is an immutable (unchangeable) sequence type used to store multiple items in a single variable.
It is similar to a list, but once created, its elements cannot be modified (no adding, removing, or changing values).

```python
mydata=(10,20,5,4,22)
```

```python
print(type(mydata))
```

## Set ##
- In Python, a set is a built-in data type used to store unordered, unique elements.
It is useful when you need to eliminate duplicates or perform mathematical set operations like union, intersection, and difference. 

```python
dataduplicates={20,20,30,30,50,4,4,50,60}
```

```python
print(dataduplicates)
```

```python
print(type(dataduplicates))
```
