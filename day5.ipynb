# Bitwise Operators #

```python
# Binary Example
a = 5  # → 0101
b = 3   #→ 0011
```

## And ##
#### 1 only if both bits are 1 

```python
a = 5
b = 3
print(a & b)   # Output: 1
```

## Or ##
#### 1  if any of  both bits are 1 

```python
print(a | b)   # Output: 7
```

# XOR (^)
- if both bit is true return 0 else 1

```python
print(a ^ b) 
```

# Right Shift (>>)

### 👉 Shift bits right (divide by 2)

```python
print(5 >> 1) 
```
