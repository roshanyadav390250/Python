# Recursion
 Recursion in Python means a function calling itself to solve a problem. It is used when a problem can be broken into smaller subproblems of the same type.

## Basic idea

A recursive function has two main parts:
Base case → condition where the function stops calling itself
Recursive case → the function calls itself with a smaller input
If you don’t define a proper base case, the function will run forever and cause a stack overflow error.
Example 1: Factorial using recursion
Factorial of a number n means:
 n! = n × (n-1) × (n-2) × ... × 1

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
def factorial(n):
   if n == 1:   # base case
       return 1
   else:
       return n * factorial(n - 1)   # recursive call

print(factorial(4))
```

# Fibonacci series


```python
def fibonacci(n):
   if n == 0:   # base case
       return 0
   elif n == 1: # base case
       return 1
   else:
       return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(10)
```
