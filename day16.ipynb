## examples of exception

```python
try:
      order=int(input("give your order please 1-4"))
      curd=10
      samosa=15
      saahi_paneer=10
      tandoori_chicken=5
      match order:
         case 1:
            print('your order: curd')
            curd-=1
            print(curd)
         case 2:
            print('your order: samosa')
            samosa-=1
            print(samosa)
         case 3:
            print('your order: saahi paneer')
            saahi_paneer-=1
            print(saahi_paneer)
         case 4:
            print('your order: tandoori chicken')
            tandoori_chicken-=1
            print(tandoori_chicken)
         case _:
            print('invalid order')
except ValueError:
      print("enter valid input")
finally          :
      print("excecution completed")
```
