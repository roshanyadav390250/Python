# Nested loops #

```python
for i in range(5):
    for j in range(4):
        print(i,j)
```

```python
for i in range(1):
    for k in range(2):
        for j in range(3):
            print(i,k,j)
```

# loops controlling 
## Breaks
- loops break immediatly

```python
for i in range(10):
    if i==5:
        break 
    print(i)
```

## Continues ##

```python
for i in range(15):
    if i==10:
        continue
    print(i)
```

## Pass ##

```python
for i in range(10):
    if i==5:
        pass
    print(i)
```
