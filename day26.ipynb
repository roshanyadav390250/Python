```python
sales = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 130],
    'Region': ['East', 'West', 'East', 'West']
})

pivot = pd.pivot_table(sales, values='Sales', index='Product', columns='Region', aggfunc='sum', fill_value=0)
print("Pivot Table (Sales by Product and Region):\n", pivot)
```

```python
sales_data = pd.DataFrame({
    'id' :[101,102,103,104],
    'Name' :['Roshan','vijay','suresh','Ramesh'],
    'Product' :['mouse','keyboard','CPU','CPU'],
    'Quantity' :[10,6,20,10],
    'Rate' :[150,200,550,800],
})
```

```python
sales_data
```

```python
sales_data['total_amount'] = sales_data['Rate']*sales_data['Quantity']
```

```python
sales_data
```

```python
pivot = pd.pivot_table(
    sales_data,
    values='Quantity',
    index='Name',
    columns='Product',
    aggfunc='sum',
    margins='True' 
)
```
