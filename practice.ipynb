# while loops

```python
i=1
while i<=5:
    print(i)
    i+=1
```

```python
i=2
while i<=20:
    print(i)
    i=i*2
```

```python
biscuits=5
while biscuits>0:
    print("biscuits hai:",biscuits)
    biscuits-=1

print("biscuits khatam!")
```

```python
lives = 3

while lives > 0:
    print("You are playing... Lives left:", lives)
    lives = lives - 1

print("Game Over!")
```
