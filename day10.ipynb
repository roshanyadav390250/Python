# ternary operaters
- short form of if-else
- The ternary operator in python is a one-line shortcut for writing simple
if -else conditions


```python
age=18
```

```python
result=print('can vote')if age>=18 else print("can't vote")
```

```python
result= 'adult' if age>=18 else 'minor'
```
