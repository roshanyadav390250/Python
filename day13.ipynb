# Match Case #

```python
vote=int(input("please enter your vote"))
bjp=100
sp=50
bsp=34
congress=80
match vote:
    case 1:
        print('you voted for bjp')
        bjp+=1
        print(bjp)
    case 2:
        print('you voted for sp')
        sp+=1
    case 3:
        print('you voted for bsp')
        bsp+=1
    case 4:
        print('you voted for congress')
        congress+=1
    case _:
        print('invalid your vote')
```

## restaurant match case ##

```python
order=int(input("give your order please 1-4"))
curd=10
samosa=15
saahi_paneer=10
tandoori_chicken=5
match order:
    case 1:
        print('your order: curd')
        curd-=1
        print(curd)
    case 2:
        print('your order: samosa')
        samosa-=1
        print(samosa)
    case 3:
        print('your order: saahi paneer')
        saahi_paneer-=1
        print(saahi_paneer)
    case 4:
        print('your order: tandoori chicken')
        tandoori_chicken-=1
        print(tandoori_chicken)
    case _:
        print('invalid order')
```
