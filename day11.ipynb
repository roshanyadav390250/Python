# while loops
- Runs While condition is true 
- The while loop runs as long as a given condition is True

```python
i=0

while i<=10:
    print(i)
    i+=1
```

```python
i=10
while i>=1:
    print(i)
    i-=1
```

```python
signal=input('enter your signal red,yellow,green,blue')
while signal=='red':
    print('stop')
    break

    
```
