```python
# Step 1: Load CSV data
df = pd.read_csv("student.csv")

# Step 2: Data overview
print(df.head())        # पहली 5 rows
print(df.info())        # columns और data types
print(df.describe())    # numerical summary (mean, std, min, max)

# Step 3: Basic Analysis
print("Total Students:", len(df))
print("Average Age:", df["age"].mean())
print("Average Score:", df["score"].mean())
print("Highest Score:", df["score"].max())
print("Lowest Score:", df["score"].min())

# Step 4: Grouping by City
city_group = df.groupby("city")["score"].mean()
print(city_group)

# Step 5: Sorting
sorted_scores = df.sort_values(by="score", ascending=False)
print(sorted_scores)

# Step 6: Visualization
# Bar chart of scores by student
df.plot(kind="bar", x="name", y="score", title="Student Scores")
plt.show()

# Pie chart of average score by city
city_group.plot(kind="pie", autopct="%1.1f%%", title="Average Score by City")
plt.show()
```
