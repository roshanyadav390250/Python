## Searching

```python
l1=[20,30,40,50]
if 20 in l1:
    print("20 is found")
else:
    print("element not found")
```

```python
for i in range(len(l1)-1):
        print(i)
```

```python
l2=[10,10,20,40,30,40,80,80,50,60,100,100,44,44]
```

```python
duplicates=set()
seen=set()

for item in l2:
    if item in seen:
        duplicates.add(item)
    else:
        seen.add(item)
print(list(duplicates))
```
