## List Comprehension ##

```python

list1 = [l for l in range(10)]
print(list1)
```

```python

squers = [l**2 for l in range(10)]
print(squers)
```

```python

cubes = [l**3 for l in range(10)]
print(cubes)
```

```python
even_number= [x for x in range(50) if x % 2==0]
print(even_number)
```

```python
marks = [23,22,46,76,78,44]
result = ["even" if x % 2 == 0 else "odd" in marks for x in marks ]
print(result)
```

```python
name = [ 'roshan,prashant,netika']
upper_name = [ name.upper() for name in name]
print(upper_name)
        
```

```python
list2 = [0,1,2,3,4,5]
l = [l**2 for l in list2]
print(l)
```

```python
listname = ['roshan,netika,prashant,vikas']
uppername= [listname.upper() for listname in listname]
print(uppername)
```

```python
lowername = [listname.lower() for listname in listname]
print(lowername)
```
